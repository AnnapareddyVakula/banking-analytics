```python
import pandas as pd
import os

print("Libraries loaded ✅")

```

```python
raw_path = '../data/raw/'

files = os.listdir(raw_path)
for f in files:
    size = os.path.getsize(raw_path + f) / (1024**3)
    print(f"{f} — {size:.2f} GB")
    
```

```python
print("Loading accepted loans... this will take 2-3 minutes ⏳")

chunks = []
chunk_size = 100000

for chunk in pd.read_csv(
    raw_path + 'accepted_2007_to_2018Q4.csv',
    chunksize=chunk_size,
    low_memory=False
):
    chunks.append(chunk)

df = pd.concat(chunks, ignore_index=True)

print(f"Done ✅")
print(f"Total rows: {len(df):,}")
print(f"Total columns: {len(df.columns)}")
```

```python
df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%Y')

year_counts = df['issue_d'].dt.year.value_counts().sort_index()
print("Loans per year:\n")
print(year_counts)
```

```python
# Filter to 2017-2018 for manageability
df_working = df[df['issue_d'].dt.year >= 2017].copy()

print(f"Full dataset:               {len(df):,} rows")
print(f"Working dataset (2017-2018): {len(df_working):,} rows")
print(f"Columns:                    {len(df_working.columns)}")
```

```python
save_path = '../data/raw/lending_club_working.csv'

df_working.to_csv(save_path, index=False)

print(f"Working file saved ✅")
print(f"Location: {save_path}")
```
